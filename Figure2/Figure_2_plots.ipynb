```python
import glob
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data for low activity threshold
for file in glob.glob("mnist_probmeta_4.52.51.json"):
    f = open(file)
    data_rig = json.load(f)
rig_Acc = data_rig["Acc"]
rig_Acc_mean = np.mean(rig_Acc, axis =2) # the accuracies are averaged over 5 runs
```

```python
# data for optimized activity threshold
for file in glob.glob("mnist_probmeta_5.53.251.75.json"):
    f = open(file)
    data_med = json.load(f)
med_Acc = data_med["Acc"]
med_Acc_mean = np.mean(med_Acc, axis =2) # the accuracies are averaged over 5 runs
```

```python
# data for high activity threshold
for file in glob.glob("mnist_probmeta_6.542.5.json"):
    f = open(file)
    data_plas = json.load(f)
plas_Acc = data_plas["Acc"]
plas_Acc_mean = np.mean(plas_Acc, axis =2) # the accuracies are averaged over 5 runs
```

```python
# data for no metaplasticity
for file in glob.glob("mnist_probmeta_100100100.json"):
    f = open(file)
    data_ref = json.load(f)
ref_Acc = data_ref["Acc"]
ref_Acc_mean = np.mean(ref_Acc, axis =2) # the accuracies are averaged over 5 runs
```

```python
task_seq = [1,2,3,4,5]

plt.plot(task_seq, med_Acc_mean[0,:]*100, linewidth = 3, color = "#648FFF", marker = ".", markersize = 20, label = "Task 1")
plt.plot(task_seq, med_Acc_mean[1,:]*100, linewidth = 3,color = "#785EF0", marker = ".", markersize = 20, label = "Task 2")
plt.plot(task_seq, med_Acc_mean[2,:]*100, linewidth = 3,color = "#DC267F", marker = ".", markersize = 20, label = "Task 3")
plt.plot(task_seq, med_Acc_mean[3,:]*100, linewidth = 3,color = "#F66102", marker = ".", markersize = 20, label = "Task 4")
plt.plot(task_seq, med_Acc_mean[4,:]*100, linewidth = 3,color = "#F8B001", marker = ".", markersize = 20, label = "Task 5")
legend = plt.legend(loc="lower left", fontsize = 'x-large')
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((1, 1, 1, 0.6))
plt.xticks( [1,2,3,4,5],fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("# Task",fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)
plt.title("Optimized Activity Threshold", fontsize = 14)
plt.savefig("optimized_threshold.eps", dpi = 300 )
```
