```python
# this file reads the energy per sample data 
#plots figure 5a,c, and d
import glob
import json
import numpy as np
import numpy.matlib

import matplotlib.pyplot as plt
import gzip
```

```python

for file in glob.glob("acc_mnist_2k*.json"):
    f = open(file)
    acc_2k = json.load(f)
for file in glob.glob("acc_mnist_8k*.json"):
    f = open(file)
    acc_8k = json.load(f)
for file in glob.glob("acc_mnist_25k*.json"):
    f = open(file)
    acc_25k = json.load(f)
for file in glob.glob("acc_full.json"):
    f = open(file)
    acc_full = json.load(f)

for file in glob.glob("prob_mnist_2k*.json"):
    f = open(file)
    prob_2k = json.load(f)
for file in glob.glob("prob_mnist_8k*.json"):
    f = open(file)
    prob_8k = json.load(f)
for file in glob.glob("prob_mnist_25k*.json"):
    f = open(file)
    prob_25k = json.load(f)
for file in glob.glob("prob_mnist_full.json"):
    f = open(file)
    prob_full = json.load(f)

```

```python
acc_mean = []

acc_mean.append(acc_2k['cont_mean'])
acc_mean.append(acc_8k['cont_mean'])
acc_mean.append(acc_25k['cont_mean'])
acc_mean.append(acc_full['cont_mean'])

acc_std = []

acc_std.append(acc_2k['cont_std'])
acc_std.append(acc_8k['cont_std'])
acc_std.append(acc_25k['cont_std'])
acc_std.append(acc_full['cont_std'])

prob_mean = []

prob_mean.append(prob_2k['cont_mean'])
prob_mean.append(prob_8k['cont_mean'])
prob_mean.append(prob_25k['cont_mean'])
prob_mean.append(prob_full['cont_mean'])

prob_std = []

prob_std.append(prob_2k['cont_std'])
prob_std.append(prob_8k['cont_std'])
prob_std.append(prob_25k['cont_std'])
prob_std.append(prob_full['cont_std'])
```

```python
# reading accuracy data and plotting figure 5d.
# mean_mnist_shared = np.mean(np.array(mnist_data_shared["Acc"]),axis=2)[:,4]
# std_mnist_shared = np.std(np.array(mnist_data_shared["Acc"]),axis=2)[:,4]

# mean_mnist_ind = np.mean(np.array(mnist_data_ind["Acc"]),axis=2)[:,4]
# std_mnist_ind = np.std(np.array(mnist_data_ind["Acc"]),axis=2)[:,4]
R_fofig, ax = plt.subplots()
x = ["200","800","2500","6000"]
w = 6
ind = np.array([30,50,70,90])

ax.bar(ind-w ,np.array(acc_mean)*100, yerr = np.array(acc_std)*100, align='edge',color = '#ffc966', alpha=1
       , ecolor='black', capsize=2, label = "Activity-dependent Metaplasticity\nw/ Gradient Accumulation", width = w)

ax.bar(ind  ,np.array(prob_mean)*100, yerr = np.array(prob_std)*100, align='edge', color = "#8AB6A4", alpha=1
       , ecolor='black', capsize=2, label = "Probabilistic Metaplasticity", width = w)
ax.set_ylabel('Mean Accuracy across Tasks (%)', fontsize =12)
ax.set_xlabel('Average Number of Samples per Class', fontsize =12)
ax.set_ylim((60,92))
plt.xticks(ind, rotation = 90)
ax.set_xticklabels( x, rotation=0, fontsize =12)
#ax.set_title('Continual Learning Benchmark Results', fontsize =16)
ax.set_ylabel('Mean Accuracy across Tasks (%)', fontsize =12)
# ax2.set_ylabel("Parameter Overhead (x)", fontsize =16)
ax.legend(loc = 'best')
# ax2.legend(loc='best')
plt.tight_layout()
plt.savefig('Figure3.eps', dpi = 300, transparent = True)
plt.show()
```

```python
with gzip.open("mnist_prob_meta_stat", "rt") as f:
    mnist_data_ind = json.load(f)
with gzip.open("mnist_shared_m_stat", "rt") as f:
    mnist_data_shared = json.load(f)
```

```python
## reading the m and w parameters
m_in_rec_shared = np.array(mnist_data_shared["m_in_rec"])
w_in_rec_shared  = np.array(mnist_data_shared["w_in_rec"])
m_in_rec_ind = np.array(mnist_data_ind["m_in_rec"])
w_in_rec_ind  = np.array(mnist_data_ind["w_in_rec"])
```

```python

n_in = 784
temp_m = m_in_rec_shared
temp_m = np.matlib.repmat(np.reshape(temp_m, (len(temp_m),1)),1, n_in)
mw_shared = temp_m*abs(w_in_rec_shared[:,:])
mw_shared = np.reshape(mw_shared, [mw_shared.shape[0]*mw_shared.shape[1], 1])


mw_ind = m_in_rec_ind[:,:]*abs(w_in_rec_ind[:,:])
mw_ind = np.reshape(mw_ind, [mw_ind.shape[0]*mw_ind.shape[1], 1])
```
